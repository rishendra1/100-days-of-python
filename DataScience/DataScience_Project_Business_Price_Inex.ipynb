```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("Business_Price_Index_100rows.csv",dtype={'Series_reference':str , 'Period': float , 'Data_value': float, 'STATUS': str, 'UNITS': str , 'Subject': str , 'Group': str , 'Series_title_1': str , 'Series_title_2': str , 'Series_title_3': str},low_memory = False)
```

```python
# set option helps to remove the
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df.drop(['STATUS', 'UNITS','Subject','Group','Unnamed: 8'],axis=1, inplace=True)
```

```python
df
```

```python
#Renaming columns
df.rename(columns={'Data_value':'Value','Series_reference':'Reference'},inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
x = df['Period']
y = df['Value']
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.bar(x,y)
plt.title('Business Price Index from 1995 to 2015')
plt.xlabel('<--Period-->')
plt.ylabel('<--Data Value-->')
plt.plot(x,y,marker='o',color='black')
plt.show()
```

```python
Total = sum(df['Value'])
print(f"Total Value of Business Index is {round(Total,2)}")
Average = Total/len(df['Value'])
print(f"{round(Average,2)} is the Average / mean Business Price Index from 1900 to 2015")
```

```python
Highest_Price = df['Value'].max()
print(f"Highest Price is {round(Highest_Price,2)}")
Lowest_Price = df['Value'].min()
print(f"Lowest Price is {round(Lowest_Price,2)}")
```

```python
Range = Highest_Price - Lowest_Price
print(f"Range of Business Price Index is {round(Range,2)}")
```
